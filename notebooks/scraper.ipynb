```python
def get_html(URL):
    try:
        html_text = requests.get(URL).text
    except Exception as e:    
        # get the exception information
        error_type, error_obj, error_info = sys.exc_info()      
        #print the link that cause the problem
        print ('ERROR FOR LINK:',URL)
        #print error info and line that threw the exception                          
        print (error_type)
        pass
    soup = BeautifulSoup(html_text, 'html.parser')
    # print(soup.prettify())
    return soup
```

### 1. Scrape content from a specific URL

Beautiful Soup transforms a complex HTML document into a complex tree of Python objects. But you’ll only ever have to deal with about four kinds of objects: Tag, NavigableString, BeautifulSoup, and Comment.

```python
URL = 'https://www.independent.co.uk/news/world/europe/berlin-protest-greece-refugees-moria-fire-migrants-b507763.html'

def get_article_content_tags(url: str):
    soup = get_html(url)
    p_tags = []
    noncontent_tags = []
    content_tags = []
    for tag in soup.findAll(True):
        if tag.name == 'p':
            p_tags.append(tag)
            try: # tags with extra attributes are non content type
                tag['class']
                noncontent_tags.append(tag)
            except KeyError:
                content_tags.append(tag)
                pass
    return content_tags, noncontent_tags, p_tags

content_tags,_,_ = get_article_content_tags(URL)
content_tags
```

### 2. Scrape links from a topic

```python
domain = "https://www.independent.co.uk/"
URL = 'https://www.independent.co.uk/topic/political-and-civil-unrest'

def get_article_urls(domain: str, url: str):
    article_urls = []
    soup = get_html(URL)
    tags_a = [tag for tag in [tag for tag in soup.find_all(True)] if tag.name == 'a']
    title_tags = []
    for tag in tags_a:
        try:
            if 'class' in tag.attrs and "title" in tag['class']:
                title_tags.append(tag)
        except KeyError:
            pass
    for tag in title_tags:
        article_urls.append(domain + tag['href'])
    return article_urls

get_article_urls(domain=domain, url = URL)
```

```python
def get_html(URL):
    try:
        html_text = requests.get(URL, verify=False).text
        soup = BeautifulSoup(html_text, 'html.parser')
        return soup
    except Exception as e:    
        # get the exception information
        error_type, error_obj, error_info = sys.exc_info()      
        #print the link that cause the problem
        print ('ERROR FOR LINK:',URL)
        #print error info and line that threw the exception                          
        print (error_type)
        pass
    # soup = BeautifulSoup(html_text, 'html.parser')
    # print(soup.prettify())
    # return soup

def get_article_urls(domain: str, url: str):
    article_urls = []
    soup = get_html(URL)
    tags_a = [tag for tag in [tag for tag in soup.find_all(True)] if tag.name == 'a']
    title_tags = []
    for tag in tags_a:
        try:
            if 'class' in tag.attrs and "title" in tag['class']:
                title_tags.append(tag)
        except KeyError:
            pass
    for tag in title_tags:
        article_urls.append(domain + tag['href'])
    return article_urls



def get_article_content_tags(url: str):
    soup = get_html(url)
    p_tags = []
    noncontent_tags = []
    content_tags = []
    for tag in soup.findAll(True):
        if tag.name == 'p':
            p_tags.append(tag)
            try: # tags with extra attributes are non content type
                tag['class']
                noncontent_tags.append(tag)
            except KeyError:
                content_tags.append(tag)
                pass
    return content_tags, noncontent_tags, p_tags
```

```python
domain = "https://www.independent.co.uk"
URL = 'https://www.independent.co.uk/topic/political-and-civil-unrest'

results = {"url":[], "content":[]}
article_urls = get_article_urls(domain=domain, url = URL)
for i in range(len(article_urls)-1):
    results["url"].append(article_urls[i])
    content_tags,_,_ = get_article_content_tags(article_urls[i])
    results["content"].append(content_tags)
```
